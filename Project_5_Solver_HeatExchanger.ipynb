```python
import ansys.fluent.core as pyfluent
from matplotlib import pyplot as plt
from matplotlib import image
%matplotlib inline
```

Fluent in solver mode

```python
solver = pyfluent.launch_fluent (precision = "double", processor_count = 4, mode = "solver", show_gui = True)
```

Read the mesh file

```python
solver.file.read_mesh(file_name = "Project_5_HeatExchanger.msh.h5")
```

```python
solver.tui.define.units("length","mm")
solver.mesh.check()
```

Check the mesh quality

```python
solver.mesh.quality()
```

```python
solver.setup.models.energy()
```

```python
solver.setup.models.energy.enabled = True
```

```python
solver.setup.models.viscous()
```

```python
solver.setup.materials()
```

To list the material from database

```python
solver.setup.materials.database.list_materials()
```

to copy material from database

```python
solver.setup.materials.database.copy_by_name(type = 'fluid', name = 'water-liquid')
```

```python
solver.setup.materials.list_materials()
```

Adding water-liquid in fluid inner and air in fluid_outer

```python
solver.setup.cell_zone_conditions.fluid()
```

```python
solver.setup.cell_zone_conditions.fluid['fluid_outer-solid'].material = 'air'
solver.setup.cell_zone_conditions.fluid['fluid_inner'].material = 'water-liquid'
```

```python
solver.setup.cell_zone_conditions.fluid()
```

```python
#now the two fluids has been added in their respective regions, checking the material of solid
solver.setup.cell_zone_conditions.solid()
```

Check the boundary conditions..

```python
solver.setup.boundary_conditions.velocity_inlet()
```

```python
# setting the velocity magnitude as 10 m/s for air, 5 m/s for water, it will be later changed for doing parametric analysis
solver.setup.boundary_conditions.velocity_inlet['air_inlet'](vmag = 10 , t = 300)
solver.setup.boundary_conditions.velocity_inlet['water_inlet'](vmag = 5 , t = 500)
```

```python
solver.setup.boundary_conditions.velocity_inlet()
```

```python
solver.setup.boundary_conditions.pressure_outlet()
```

```python
solver.solution.initialization.hybrid_initialize()
```

```python
solver.solution.run_calculation.iter_count = 500
```

```python
solver.solution.run_calculation.iterate()
```

```python
solver.file.write(file_type = 'case-data', file_name = "Project_5_HeatExchanger.cas.h5")
```

```python
solver.exit()
```
